# Using pantherdb to get the orthologs from ref proteom

```python
import pandas as pd
import Bio
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
with open('../Results/matches2refproteom.pickle', 'rb') as f:
    uniprot2seq = pickle.load(f)
```

# Retrieving the orthologs

```python
orthologs_table = pd.read_csv('../Data/CANLF_HUMAN_all.tab',sep='\t',header=None)

orthologs_table.head()
```

```python
d = {}
for i in orthologs_table.index:
    if "HUMAN" in orthologs_table[0][i]:
        for el in orthologs_table[0][i].split("|"):
            if "UniProtKB" in el:
                human = el.split("UniProtKB=")[1]
        for el in orthologs_table[1][i].split("|"):
            if "UniProtKB" in el:
                canlf = el.split("UniProtKB=")[1]
    else:
        for el in orthologs_table[1][i].split("|"):
            if "UniProtKB" in el:
                human = el.split("UniProtKB=")[1]
        for el in orthologs_table[0][i].split("|"):
            if "UniProtKB" in el:
                canlf = el.split("UniProtKB=")[1]
    store = (human,orthologs_table[2][i])
    if canlf in d:
        d[canlf].append(store)
    else:
        d[canlf] = [store]
```

```python
len(d)
```

```python
d["A0A5F4CFQ7"]
```

# Matching from old uniprot id to orthologs 

```python
orthologs_found = {}
orthologs_not_found = {}

for key, values in uniprot2seq.items():
    found = False
    for value in values:
        canlf_ref_id = value.name.split("|")[1]
        if canlf_ref_id not in d:
            continue
        else:
            human_ref_id = ""
            b=True
            for el in d[canlf_ref_id]:
                if el[1] == 'LDO':
                    human_ref_id = el[0]
                    b=False
                    break
                else:
                    human_ref_id = el[0]
            if b :
                print(key,canlf_ref_id,human_ref_id)
            orthologs_found[key] = human_ref_id
            found = True
            break
            
    if not found :
        orthologs_not_found[key] = value
```

# lets check how many lines we get from the filtered set of proteins

```python
filtered_prots = pd.read_csv("../Data/proteinGroups_Filtered.txt", sep="\t")
filtered_prots
```

```python
filtered_prots["Majority protein IDs"][1303]
```

```python
list_of_found_indices = []
list_of_not_found_indices = []
for i in filtered_prots.index:
    s = filtered_prots["Majority protein IDs"][i]
    l = s.split(";")
    for el in l:
        if el in orthologs_found:
            list_of_found_indices.append(i)
            break
    if list_of_found_indices[-1] != i:
        list_of_not_found_indices.append(i)
```

```python
found_orthologs_table = pd.DataFrame([[key,val] for key, val in orthologs_found.items()],columns= ["CANLF_UniProtKB","HUMAN_UniprotKB"])
found_orthologs_table
```

```python
not_found_orthologs_table = pd.DataFrame([[key,np.nan] for key, val in orthologs_not_found.items()],columns= ["CANLF_UniProtKB","HUMAN_UniprotKB"])
not_found_orthologs_table
```

```python
matches = pd.concat([found_orthologs_table,not_found_orthologs_table])
matches
```
