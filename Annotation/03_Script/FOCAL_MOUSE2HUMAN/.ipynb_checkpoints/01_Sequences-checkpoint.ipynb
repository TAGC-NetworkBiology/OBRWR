# Retrieving sequences from uniprot/uniparc CANLF proteins

```python
from Bio import SeqIO
import Bio
import pandas as pd
import numpy as np
```

```python
experiment_name = "CANLF2HUMAN"
back_path = "../../"
uniprot_file_name = "CANLF_uniprot.fasta"
uniparc_file_name_fasta = "uniparc-organism-Canis+lupus+familiaris+(Dog)+(Canis+familiaris)+%5--.fasta"
uniparc_file_name_tab = "uniparc-compressed_true_download_true_fields_upi_2Caccession_2Cseque-2022.11.18-11.21.10.63.tsv"
exp_protein_file = "proteinGroups_Filtered.txt"
```

```python
with open(back_path + "01_Reference/"+experiment_name+"/"+uniprot_file_name) as handle:
    uniprot_CANLF = SeqIO.to_dict(SeqIO.parse(handle, "fasta"))
    
with open(back_path+"01_Reference/"+experiment_name+"/"+uniparc_file_name_fasta) as handle:
    uniparc_CANLF_dict = SeqIO.to_dict(SeqIO.parse(handle, "fasta"))

uniparc_CANLF_tab = pd.read_csv(back_path+"01_Reference/"+experiment_name + "/" + uniparc_file_name_tab,sep='\t')
```

```python
uniparc_CANLF_tab.head()
```

```python
df = pd.read_csv(back_path+"00_InputData/"+experiment_name+"/"+exp_protein_file,sep='\t')
df.head()
```

```python
prot_in_study = {}
for s in df["Majority protein IDs"]:
    for prot in s.split(";"):
        prot_in_study[prot] = Bio.SeqRecord.SeqRecord('')
```

```python
print(len(prot_in_study.keys()))
len(prot_in_study.keys()) == len(np.unique(list(prot_in_study.keys())))
```

```python
len(uniprot_CANLF)
```

# Uniprot

```python
for el in uniprot_CANLF.keys():
    uniprotKB_id = el.split("|")[1]
    if uniprotKB_id in prot_in_study.keys():
        prot_in_study[uniprotKB_id] = uniprot_CANLF[el]
```

```python
prot_in_study['A0A077LQA5'].seq
```

```python
uniprot_found = {}
uniprot_not_found = {}
for key,val in prot_in_study.items():
    if  val.seq != '' :#type(val) == Bio.SeqRecord.SeqRecord:
        uniprot_found[key] = val
    else:
        uniprot_not_found[key] = []
```

# Uniparc

```python
 - uniparc_CANLF_tab["UniProtKB"].isna()
```

```python
uniparc_CANLF_tab_id = uniparc_CANLF_tab[- uniparc_CANLF_tab["UniProtKB"].isna()]
uniparc_CANLF_tab_id
```

```python
uniparc_CANLF = {}
for i in uniparc_CANLF_tab_id.index :
    for prot in uniparc_CANLF_tab_id['UniProtKB'][i].split(";"):
        prot_id = prot.split(".")[0].strip()
        tmp_seq = Bio.SeqRecord.SeqRecord(id = prot_id,seq = Bio.Seq.Seq(uniparc_CANLF_tab_id['Sequence'][i]))
        uniparc_CANLF[prot_id] = tmp_seq
```

```python
uniparc_found = {}
uniparc_not_found = {}
for key in uniprot_not_found.keys():
    if key in uniparc_CANLF:
        uniparc_found[key] = uniparc_CANLF[key]
    else:
        uniparc_not_found[key] = []
```

```python
for el in uniparc_not_found.keys():
    print(el)
```

# Saving as pickle and as fasta

```python
uniprot_found.update(uniparc_found)
```

```python
len(uniprot_found)
```

```python
import pickle
```

```python
with open(back_path+"05_Output/"+experiment_name+"/"+experiment_name+"_sequences.pickle", 'wb') as f:
    pickle.dump(uniprot_found, f, pickle.HIGHEST_PROTOCOL)
```

```python
with open(back_path+"05_Output/"+experiment_name+"/"+experiment_name+"_sequences.fasta", "w") as output_handle:
    SeqIO.write( uniprot_found.values(),output_handle, "fasta")
```
