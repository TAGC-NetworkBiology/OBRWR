```python
import pandas as pd
import numpy as np
```

```python
back_path="../"
matches_file_name = "CANLF2HUMAN_matches_annotated.tsv"
sen_CIBN_file = "os_sensitive_CIBN.tsv"
sen_OS_file = "os_sensitive_OS.tsv"
```

```python
matches = pd.read_csv(back_path+"00_InputData/"+matches_file_name,sep='\t')
sensitive_CIBN = pd.read_csv(back_path+"00_InputData/"+sen_CIBN_file,sep='\t')
sensitive_OS = pd.read_csv(back_path+"00_InputData/"+sen_OS_file,sep='\t')
```

```python
d = {}
invd = {}
for i in range(len(matches)):
    canlf_prot = matches['CANLF_UniProtKB'][i]
    human_prot = matches['HUMAN_UniProtKB_ACC'][i]
    if canlf_prot not in d.keys():
        d[canlf_prot] = [human_prot]
    else:
        d[canlf_prot].append[human_prot]
    if human_prot not in invd.keys():
        invd[human_prot] = [canlf_prot]
    else:
        invd[human_prot].append(canlf_prot)
matches.head()
```

```python
sensitive_CIBN.head()
```

```python
CIBNt0 = np.mean(sensitive_CIBN[sensitive_CIBN.columns[2:5]][1:],axis=1)
CIBNt5 = np.mean(sensitive_CIBN[sensitive_CIBN.columns[5:8]][1:],axis=1)
CIBNt20 = np.mean(sensitive_CIBN[sensitive_CIBN.columns[8:11]][1:],axis=1)
vals = [CIBNt5 - CIBNt0, CIBNt20 - CIBNt0, CIBNt20 - CIBNt5]
dfsensitiveCIBNval = sensitive_CIBN[["Entry"]][1:]
dfsensitiveCIBNval.loc[:,("ACC")] = dfsensitiveCIBNval.loc[:,("Entry")].map({el:";".join(val) for el,val in d.items() if np.nan not in val})
for col_name,val in zip(sensitive_CIBN.columns[-3:],vals):
    dfsensitiveCIBNval.loc[:,(col_name)] = sensitive_CIBN[col_name]*np.abs(val)
```

```python
OSt0 = np.mean(sensitive_OS[sensitive_OS.columns[11:14]][1:],axis=1)
OSt5 = np.mean(sensitive_OS[sensitive_OS.columns[14:17]][1:],axis=1)
OSt20 = np.mean(sensitive_OS[sensitive_OS.columns[17:20]][1:],axis=1)
vals = [OSt5-OSt0, OSt20-OSt0,OSt20-OSt5]
dfsensitiveOSval = sensitive_OS[["Entry"]][1:]
dfsensitiveOSval.loc[:,("ACC")] = dfsensitiveOSval.loc[:,("Entry")].map({el:";".join(val) for el,val in d.items() if np.nan not in val})
for col_name,val in zip(sensitive_OS.columns[-3:],vals):
    dfsensitiveOSval.loc[:,(col_name)] = sensitive_OS[col_name]*np.abs(val)
```

```python
dfsensitiveCIBNval.to_csv(back_path+"05_Output/"+"CIBN/os_sensitive_CIBN_withval.tsv",sep='\t',index=False)
dfsensitiveOSval.to_csv(back_path+"05_Output/"+"OS/os_sensitive_OS_withval.tsv",sep='\t',index=False)
```
