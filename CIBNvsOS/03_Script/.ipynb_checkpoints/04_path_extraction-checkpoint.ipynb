### Functions only for simple paths (a lot of non simple path contribute to the probability of reaching a protein)
So deprecated

```python
def enumerate_paths(G,source,target,T,r=0.9,minc = 1e-3,beta=0.7,weight='WeightsStable',N=100):
    d = deque([([source],1)])
    l = []
    os = 1 
    s = 0
    while len(d) != 0 and s < r*T and np.abs(os-s) > minc*T:
        path,score = d.popleft()
        last = path[-1]
        if last == target:
            l.append((path,0.7*1e4*score))
            os = s
            s += beta*1e4*score
        for neigh in G.neighbors(last):
            escore = G.get_edge_data(last,neigh)[weight]
            d.append((path+[neigh],score*(1-beta)*escore))
    return [el[0] for el in l],s

def enumerate_sp(G,source,target,weight=lambda x,y,d : -np.log((1-0.7)*d['weights'+experiment]),r=1e-2,beta=0.7):
    os = 1e6
    s = 0
    paths = nx.shortest_simple_paths(G, source, target, weight)
    l = []
    while np.abs(os-s) > r*G.nodes[target]['Pi'+experiment]:
        path = next(paths)
        os = s 
        s +=path_weight(G,path,weight='weights'+experiment,beta=beta)
        l.append(path)
    return l,s

def enumerate_sp2(G,source,target,weight=lambda x,y,d : -np.log((1-0.7)*d['weights'+experiment]),r=1-2e-1):
    os = 1e4
    s = 0
    T = G.nodes[target]['Pi'+experiment]
    paths = nx.shortest_path(G, source, target, weight)
    l = []
    print(r)
    while s/T < r:
        path = next(paths)
        os = s 
        s +=path_weight(G,path,weight='weights'+experiment)
        l.append(path)
    return l,s

def path_weight(G, path,weight,tot=1e4,beta=0.7):
    cost=1
    if not nx.is_path(G, path):
        raise nx.NetworkXNoPath("path does not exist")
    for node, nbr in nx.utils.pairwise(path):
        d = G.get_edge_data(node,nbr)
        cost *= (1-beta)*d[weight]
    return beta*tot*cost

from itertools import islice
def k_shortest_paths(G, source, target, k, weight=None):
    return list(
        islice(nx.shortest_simple_paths(G, source, target, weight=weight), k)
    )
```

### Functions for paths with loops 

```python
from heapq import *

def to_probas(distances):
    for node1 in distances.keys():
        for node2 in distances[node1].keys():
            distances[node1][node2] = np.exp(-distances[node1][node2])

def maxprob_to_targets(node,targets_left,sp_distances,Ts):
    maxprob = 0
    for target in targets_left:
        d = sp_distances[node][target]
        if d > maxprob:
            maxprob = d
    return maxprob
    
def enumerate_path_heap(G,targets,experiment,source='SRC_HUMAN',beta=0.7,r=1e-1,Tot=1e4,verbose=False,mode='total'):
    weight_sp = lambda u,v,d: -np.log((1-beta)*d['weights'+experiment])
    Ts = {t:G.nodes[t]['Pi'+experiment] for t in targets}
    pathstotarget = {t:[[],0] for t in targets}
    sp_distances = dict(nx.all_pairs_dijkstra_path_length(G,weight = weight_sp))
    to_probas(sp_distances)
    if verbose :
        print('Ts :',Ts)
    weight = lambda d: d['weights'+experiment]
    
    initial_item = [-beta*Tot,beta*Tot,[source]]
    h = []
    heappush(h,initial_item)
    
    targets_left = set(targets)
    while len(targets_left) and len(h):
        dist2target, path_prob, path = heappop(h)
        if path[-1] in targets_left:
            tar = path[-1]
            pathstotarget[tar][0].append(path)
            pathstotarget[tar][1] += path_prob
            tmp_b = False
            if mode == 'percent' and path_prob/Ts[tar] < r:
                targets_left.remove(tar)
                tmp_b = True
            elif mode == 'total' and pathstotarget[tar][1] > (1-r)*Ts[tar]:
                targets_left.remove(tar)
                tmp_b = True
            if tmp_b and verbose:
                print('Target found :',tar)
                print('Probas found :',{t:pathstotarget[t][1] for t in targets})
                print('targets_left :', targets_left)
        for neigh in nx.neighbors(G,path[-1]):
            wedge = (1-beta)*weight(G.edges[(path[-1],neigh)])
            maxp = maxprob_to_targets(neigh,targets_left,sp_distances,Ts)
            heappush(h,[-path_prob*wedge*maxp,path_prob*wedge,path+[neigh]])
    return pathstotarget
```

## Time 5 - 0

```python
back_path = "../"
experiment = "CIBN"
time_exp = 'B5 - B0'
folder_exp = back_path+'05_Output/'+experiment+"/"
folder = folder_exp + "random_"+time_exp+'/' 
subdirectories = [f.path for f in os.scandir(folder) if f.is_dir() and (f.name[0] != '.')]
OriginalG = nx.read_gml(folder_exp+'signor_subgraph_'+experiment+time_exp+'.gml')
```

```python
targets_attr = nx.get_node_attributes(OriginalG,'Targets')
original_targets = [n for n in OriginalG.nodes() if targets_attr[n]]
```

```python
original_targets
```

```python
np.sum(list(nx.get_node_attributes(OriginalG,'Pi'+experiment).values()))
```

Retrieving the paths which account for 80% of reaching probability.

Checking and then plotting the cumulative probabilities, ranking paths by probability.

```python
pathstotarget = enumerate_path_heap(OriginalG,original_targets,experiment,r=1e-2,mode='percent')
```

```python
paths = []
for target in original_targets:
    T = OriginalG.nodes[target]['Pi'+experiment]
    tpaths,s = pathstotarget[target]
    paths += tpaths
    print(f'Explained prob. of {target}: {s/T}')
    print(f'Number of paths : {len(tpaths)}')
```

```python
N = len(original_targets)
nrows = int(np.ceil(N/3))
f,axs = plt.subplots(nrows,3,figsize=(21,nrows*5))
for i in range(len(original_targets)):
    target = original_targets[i]
    ax = axs[i//3][i%3]
    if not i%3:
        ax.set_ylabel('Cumulative probabilities')
    ax.set_title(target)
    ax.set_xlabel('Ranked paths')
    
    T = OriginalG.nodes[target]['Pi'+experiment]
    paths4viz = pathstotarget[target][0]
    probas = [path_weight(OriginalG,path,weight='weights'+experiment)/T for path in paths4viz]
    cumsum_prob = list(np.cumsum(probas))
    ax.scatter(range(len(paths4viz)+1),[0]+cumsum_prob)
    ax.plot([-1,len(paths4viz)+1],[0,0], ls = '--',c='grey')
    ax.plot([0,0],[-0.2,1.2], ls = '--',c='grey')
    ax.set_xlim(-0.5,len(paths4viz)+0.5)
    ax.set_ylim(-0.1,1.1)
```

```python
BW = []
OW = []
for path in paths:
    print("Path :",path)
    bw = path_weight(OriginalG,path,weight='weights'+experiment)
    ow = path_weight(OriginalG,path,weight="WeightsStable")
    print("Biased Weights :",bw)
    BW.append(bw)
    OW.append(ow)
    T = OriginalG.nodes[path[-1]]['Pi'+experiment]
    st = OriginalG.nodes[path[-1]]['Stable']
    print("Original Weights :",ow)
    print("DeltaWeights :",bw-ow)
    print(f'Delta in percent explained :{(bw-ow)*100/(T-st)}%')
```

```python
f,ax = plt.subplots(1,1,figsize=(14,10))
target4viz = 'PRAG1_HUMAN'

ax.set_xlabel('Ranked paths',fontsize=18)
ax.set_title(target4viz,fontsize=22)
ax.set_ylabel('Cumulative probabilities',fontsize=18)

T = OriginalG.nodes[target4viz]['Pi'+experiment]
paths4viz = pathstotarget[target4viz][0]
probas = [path_weight(OriginalG,path,weight='weights'+experiment)/T for path in paths4viz]
cumsum_prob = [0]+list(np.cumsum(probas))
ax.scatter(range(len(paths4viz)+1),cumsum_prob,s=60)
ax.plot([-1,len(paths4viz)+1],[0,0], ls = '--',c='grey')
ax.plot([0,0],[-0.2,1.2], ls = '--',c='grey')
ax.set_xlim(-0.5,len(paths4viz)+0.5)
ax.set_ylim(-0.1,1.1)
for path,proba in zip(paths4viz,probas) :
    print(path,proba)
```

```python
pvals = np.ones(len(paths))
distribs = [[] for _ in range(len(paths))]

for subf in subdirectories:
    with open(subf+'/w_hat.pickle','rb') as f:
        What = pickle.load(f)
    with open(subf+'/inverse_mapping.pickle','rb') as f:
        inverse_mapping = pickle.load(f)
    with open(subf+'/eps_crit.pickle','rb') as f:
        eps_crit = pickle.load(f)
    pi = pd.read_csv(subf+'/pi.tsv',sep='\t',index_col=0,skiprows=1,names=['Pi'])
    # Build random graph
    tmpG = nx.from_scipy_sparse_matrix(What.T,create_using=nx.DiGraph)
    tmpG = nx.relabel_nodes(tmpG,inverse_mapping)
    # Compute paths length
    for i in range(len(pvals)):
        bw = path_weight(tmpG,paths[i],weight='weight')
        PiOS = OriginalG.nodes[paths[i][-1]]['Pi'+experiment]
        Pirandom = pi['Pi'][paths[i][-1]]
        Pistable = OriginalG.nodes[paths[i][-1]]['Stable']
        distribs[i].append(bw/Pirandom)
        
        if bw/Pirandom > 1:
            print(paths[i])
            print(f'bw : {bw}')
            print(f'original bw : {BW[i]}')
            print(f'original weight : {OW[i]}')
            print(f'eps_crit : {eps_crit}')
            print(f'Pirandom: {Pirandom}')
            print(f'Pistable: {Pistable}')
        if bw/Pirandom > BW[i]/PiOS:
            pvals[i] += 1
    # Update p-val
pvals /= len(subdirectories)+1
```

```python
back_path = "../"
experiment = "OS"
time_exp = 'A5 - A0'
folder_exp = back_path+'05_Output/'+experiment+"/"
folder = folder_exp + "random_"+time_exp+'/' 
subdirectories = [f.path for f in os.scandir(folder) if f.is_dir() and (f.name[0] != '.')]
OriginalG = nx.read_gml(folder_exp+'signor_subgraph_'+experiment+time_exp+'.gml')
```

```python
targets_attr = nx.get_node_attributes(OriginalG,'Targets')
original_targets = [n for n in OriginalG.nodes() if targets_attr[n]]
```

```python
original_targets
```

```python
np.sum(list(nx.get_node_attributes(OriginalG,'Pi'+experiment).values()))
```

Retrieving the paths which account for 80% of reaching probability.

Checking and then plotting the cumulative probabilities, ranking paths by probability.

```python
pathstotarget = enumerate_path_heap(OriginalG,original_targets,experiment,r=1e-2,mode='percent')
```

```python
paths = []
for target in original_targets:
    T = OriginalG.nodes[target]['Pi'+experiment]
    tpaths,s = pathstotarget[target]
    paths += tpaths
    print(f'Explained prob. of {target}: {s/T}')
    print(f'Number of paths : {len(tpaths)}')
```

```python
N = len(original_targets)
nrows = int(np.ceil(N/3))
f,axs = plt.subplots(nrows,3,figsize=(21,nrows*5))
for i in range(len(original_targets)):
    target = original_targets[i]
    ax = axs[i//3][i%3]
    if not i%3:
        ax.set_ylabel('Cumulative probabilities')
    ax.set_title(target)
    ax.set_xlabel('Ranked paths')
    
    T = OriginalG.nodes[target]['Pi'+experiment]
    paths4viz = pathstotarget[target][0]
    probas = [path_weight(OriginalG,path,weight='weights'+experiment)/T for path in paths4viz]
    cumsum_prob = list(np.cumsum(probas))
    ax.scatter(range(len(paths4viz)+1),[0]+cumsum_prob)
    ax.plot([-1,len(paths4viz)+1],[0,0], ls = '--',c='grey')
    ax.plot([0,0],[-0.2,1.2], ls = '--',c='grey')
    ax.set_xlim(-0.5,len(paths4viz)+0.5)
    ax.set_ylim(-0.1,1.1)
```

```python
BW = []
OW = []
for path in paths:
    print("Path :",path)
    bw = path_weight(OriginalG,path,weight='weights'+experiment)
    ow = path_weight(OriginalG,path,weight="WeightsStable")
    print("Biased Weights :",bw)
    BW.append(bw)
    OW.append(ow)
    T = OriginalG.nodes[path[-1]]['Pi'+experiment]
    st = OriginalG.nodes[path[-1]]['Stable']
    print("Original Weights :",ow)
    print("DeltaWeights :",bw-ow)
    print(f'Delta in percent explained :{(bw-ow)*100/(T-st)}%')
```

```python
f,ax = plt.subplots(1,1,figsize=(14,10))
target4viz = 'PRAG1_HUMAN'

ax.set_xlabel('Ranked paths',fontsize=18)
ax.set_title(target4viz,fontsize=22)
ax.set_ylabel('Cumulative probabilities',fontsize=18)

T = OriginalG.nodes[target4viz]['Pi'+experiment]
paths4viz = pathstotarget[target4viz][0]
probas = [path_weight(OriginalG,path,weight='weights'+experiment)/T for path in paths4viz]
cumsum_prob = [0]+list(np.cumsum(probas))
ax.scatter(range(len(paths4viz)+1),cumsum_prob,s=60)
ax.plot([-1,len(paths4viz)+1],[0,0], ls = '--',c='grey')
ax.plot([0,0],[-0.2,1.2], ls = '--',c='grey')
ax.set_xlim(-0.5,len(paths4viz)+0.5)
ax.set_ylim(-0.1,1.1)
for path,proba in zip(paths4viz,probas) :
    print(path,proba)
```

```python
pvals = np.ones(len(paths))
distribs = [[] for _ in range(len(paths))]

for subf in subdirectories:
    with open(subf+'/w_hat.pickle','rb') as f:
        What = pickle.load(f)
    with open(subf+'/inverse_mapping.pickle','rb') as f:
        inverse_mapping = pickle.load(f)
    with open(subf+'/eps_crit.pickle','rb') as f:
        eps_crit = pickle.load(f)
    pi = pd.read_csv(subf+'/pi.tsv',sep='\t',index_col=0,skiprows=1,names=['Pi'])
    # Build random graph
    tmpG = nx.from_scipy_sparse_matrix(What.T,create_using=nx.DiGraph)
    tmpG = nx.relabel_nodes(tmpG,inverse_mapping)
    # Compute paths length
    for i in range(len(pvals)):
        bw = path_weight(tmpG,paths[i],weight='weight')
        PiOS = OriginalG.nodes[paths[i][-1]]['Pi'+experiment]
        Pirandom = pi['Pi'][paths[i][-1]]
        Pistable = OriginalG.nodes[paths[i][-1]]['Stable']
        distribs[i].append(bw/Pirandom)
        
        if bw/Pirandom > 1:
            print(paths[i])
            print(f'bw : {bw}')
            print(f'original bw : {BW[i]}')
            print(f'original weight : {OW[i]}')
            print(f'eps_crit : {eps_crit}')
            print(f'Pirandom: {Pirandom}')
            print(f'Pistable: {Pistable}')
        if bw/Pirandom > BW[i]/PiOS:
            pvals[i] += 1
    # Update p-val
pvals /= len(subdirectories)+1
```
