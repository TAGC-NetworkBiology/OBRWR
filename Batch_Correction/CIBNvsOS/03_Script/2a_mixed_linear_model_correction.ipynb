```r
library(lme4)
library(sjPlot) 
library(readr)
library(lmerTest)
library(Biobase)
library(ggplot2)
library(impute)
```

```r
impute_perc_knn <- function(data){
  
  percentiles = apply(data,2,function(x)(quantile(x,seq(0,1,by=0.01),na.rm=TRUE)[2]))
  
  for (i in 1:length(data[,1])){
    for (k in 1:3){
      if (sum(is.na(data[i,(3*(k-1)+1):(3*k)])) == 3){
        data[i,(3*(k-1)+1):(3*k)] <- percentiles[(3*(k-1)+1):(3*k)]
        print(i)
      }
    }
  }
  
  imputed <- cbind(impute.knn(data[,1:3],k=3,rowmax=1)[["data"]],
                   impute.knn(data[,4:6],k=3,rowmax=1)[["data"]],
                   impute.knn(data[,7:9],k=3,rowmax=1)[["data"]])
  imputed
}

mlm_correction <- function(data){
  
  rep <- factor(c(1,2,3,1,2,3,1,2,3))
  time <- factor(c(1,1,1,2,2,2,3,3,3))
  
  pr1 <- data.frame(PROT = factor(rep(as.character( table[1,1]),9)),
                    REP = rep, TIME = time, intensity = as.vector(t(data)[,1]))
  
  for( i in 2:nrow(data)){
    pr1 <- rbind(pr1,data.frame(PROT = factor(rep(as.character( table[i,1]),9)),
                                REP = rep, TIME = time, 
                                intensity = as.vector(t(data)[,i])))
  }
  model <- lmer(intensity~ PROT:TIME+(1|PROT)+(1|PROT:REP), pr1)
  
  res_data = t(matrix(residuals(model),nrow=9))

  print(summary(model))
  effects <- ranef(model)
  pr1$nord[1] <- 0
  for (i in 1:length(pr1$nord)){
    pr1$nord[i] <- pr1$intensity[i] - effects[["PROT:REP"]][paste(pr1$PROT[i],pr1$REP[i],sep=':'),]
  }
  
  without_r <- t(matrix(pr1$nord,nrow=9))
  colnames(without_r) <- colnames(data)
  
  
  without_r
}
```

```r
file_name="3old+4old_IgGpos_Normed_Log"
output_folder="../05_Output"
table <- read_tsv(paste(output_folder,"/02_IntermediaryFiles/",file_name,".tsv",sep=""))
table_mat <- table[,3:20]


# Mixed linear model correction for random effect due to batch

data_B <- table_mat[,1:9]
corr_B <- mlm_correction(data_B)
imput_corr_B <- impute_perc_knn(corr_B)
print(imput_corr_B)
data_A <- table_mat[,10:18]
corr_A <- mlm_correction(data_A)
imput_corr_A <- impute_perc_knn(corr_A)
```

```r
boxplot(table_mat)                     
boxplot(cbind(corr_B,corr_A))
boxplot(cbind(imput_corr_B,imput_corr_A))
```

```r
full_table_imputed <- cbind(table[c("Entry","Gene_Name")],imput_corr_B,imput_corr_A)
write.table(full_table_imputed,paste(output_folder,"/02_IntermediaryFiles/",file_name,"_MLMcorrected.tsv",sep=""),sep='\t',row.names=FALSE)
```
